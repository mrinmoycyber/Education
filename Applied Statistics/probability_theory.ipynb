```python
# Counting
def factorial(n):
  f = 1
  for i in range(1, n+1):
    f = f * i
  return f
```

```python
def P(n, r):
  return int(factorial(n)/factorial(n-r))
```

```python
def C(n, r):
  return int(factorial(n)/(factorial(r) * factorial(n-r)))
```

```python
C(4, 2)
```

```python
P(5, 2)
```
