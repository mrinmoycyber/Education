Notes:

If

```python
# Counting
def factorial(n):
  f = 1
  for i in range(1, n+1):
    f = f * i
  return f
```

```python
def P(n, r):
  return int(factorial(n)/factorial(n-r))
```

```python
def C(n, r):
  return int(factorial(n)/(factorial(r) * factorial(n-r)))
```

```python
C(4, 2)
```

```python
P(5, 2)
```

```python
from itertools import permutations
from itertools import combinations
```

```python
# Get all the permutations
sequence = [1, 2, 3, 4]

for p in permutations(sequence):
  print(p)
```

```python
P(8, 3)
```

```python
sequence = [1, 2, 3, 4]

# Generate permutations of length 2
for p in permutations(sequence, 2):
    print(p)
```

```python
# Get all the combination
sequence = [1, 2, 3, 4]

for c in combinations(sequence, 3):
  print(c)
```
